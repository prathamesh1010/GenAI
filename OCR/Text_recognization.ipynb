```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
import cv2
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from keras.utils import to_categorical
import pytesseract

# Load images
img = mpimg.imread('image.jpg')
img = cv2.resize(img, (224, 224))
img = img / 255.0

# Convert the image to an 8-bit unsigned integer array
img_8bit = (img * 255).astype(np.uint8)

# Create a dataset of images
imgs = np.array([img_8bit])

# Define labels for the images
labels = np.array([0])  # Replace with actual labels

# One-hot encode the labels
labels = to_categorical(labels, num_classes=10)

# Use the entire dataset as both the training and testing sets
train_imgs = imgs
val_imgs = imgs
train_labels = labels
val_labels = labels

# Define the CNN model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(train_imgs, train_labels, epochs=10, batch_size=1, validation_data=(val_imgs, val_labels))

# Evaluate the model
loss, accuracy = model.evaluate(val_imgs, val_labels)
print(f'Model accuracy: {accuracy:.3f}')

# Use Tesseract-OCR to extract text from the image
text = pytesseract.image_to_string(img_8bit)
print(f'Extracted text: {text}')
```
